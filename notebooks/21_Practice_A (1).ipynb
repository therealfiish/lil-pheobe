Before we begin, it is important to look at the steps in analyzing real-world data. The process of doing real data science typically follows these steps:

1. **Inspect**: You should *always* look at your data before you attempt to do anything.
   a. Open up the file in Excel or another program
   b. View in Anaconda
   c. Read the datafram with Pandas.

2. **Clean**: Look for missing values ('Nan') or values out of logical range and remove these values.
   a. Can remove in the original data source. This is not recommended as this changes the data source and impacts its use for future analysis.
   b. Pandas code to remove rows and/or columns

3. **Transform**: The data often need to be modified in some way, such as changing strings to integers, reading in dates correctly, changing the units, transposing the data, etc. Depending on what you want to do, you will need to leverage a variety of code.

4. **Model**: This is where you get to present the data and determine what it is 'telling you'. In other words, what information can we determine based on the model? How can this information impact next steps and/or deision making? finally you can get to the fun part - thinking about what your data is telling you!
