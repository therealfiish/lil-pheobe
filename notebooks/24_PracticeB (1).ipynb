## 24B: Regression using NumPy

NumPy has pre-built functions to calculate a linear regression model based on the data we provide.

- `np.polyfit` You can read more here: (https://numpy.org/doc/stable/reference/generated/numpy.polyfit.html)
- `np.poly1d` You can read more here: (https://docs.scipy.org/doc/numpy-1.15.1/reference/generated/numpy.poly1d.html)

Fitting regression models with your own written functions can be a pain when there is a lot of data or you have a complicated function. NumPy makes this process very fast and efficient. Let's watch the video below for a quick tutorial! Note the similarities between `np.polyfit`/`np.poly1d` and our `lin_reg_mb`/`lin_regression` functions above.

```python
# Import the needed modules

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array of x-values

# what does this command do, specifically, what does 'seed'do?
# The seed function sets a starting point for random number generation
# Using the same seed value will produce the same sequence of random numbers every time
# This makes random experiments reproducible
np.random.seed(1) 


# What does this command do?
# np.arange(0,25,1) creates a numpy array with values from 0 to 24 (inclusive)
# with a step size of 1, generating 25 evenly spaced integers
x = np.arange(0,25,1)
```

**Try creating another random list where you identify the seed value. Do it with a person next to you. Do you get the same _random_ numbers?**

<font size=+2> &#9999; <font>  They produce same numbers when we have the same seed.

```python
# Type your code here to generate random numbers with a seed value

# Generate random numbers with the same seed
np.random.seed(7)
random_numbers = np.random.randint(0, 100, 10)
print("Random numbers with seed 7:", random_numbers)
```

```python
# What does the command below do? How does it help in what we are trying to learn?

# This command generates y-values with a linear relationship to x plus random noise
# It creates data points that follow a linear trend (y = x) with some random variation
# The normal distribution with mean 1 and standard deviation 3 adds noise to each point
# This simulates real-world data where measurements include random errors/variations

y = x + np.random.normal(1,3,25)
```

```python
plt.scatter(x,y)
plt.show()
```

### Regression with non-linear relationships

```python
# What is the line of code doing? Input your own comment to explain.
# This command fits a polynomial of degree 1 (a line) to the data points (x,y)
data_in = np.polyfit(x,y,1)
```

```python
# What is the line of code doing? Input your own comment to explain.
# This command fits a polynomial of degree 1 (a line) to the data points (x,y)
data_in
```

```python
# What are the lines of code doing? Input your own comments to explain.
# The command below extracts the slope (m) and intercept (b) of the fitted line
# The first element of data_in is the slope (m) and the second element is the intercept (b)
m_exp = data_in[0]
b_exp = data_in[1]
```

```python
# write the line of best fit based on the values from np.polyfit and name it 'output_exp'.

output_exp = m_exp * x + b_exp
```

```python
# What is the line of code doing? Input your own comment to explain.
# output_exp is the line of best fit calculated using the slope (m) and intercept (b)
output_exp
```

```python
# What are the lines of code doing? Input your own comments to explain.

# The command below creates a scatter plot of the original data points (x,y)
plt.scatter(x,y)
plt.plot(x,output_exp)
plt.show()
```

```python
# What is the line of code doing? Input your own comment to explain. What does the "2" mean?
# This command fits a polynomial of degree 2 (a quadratic curve) to the data points (x,y)
inputs_quad = np.polyfit(x,y,2)
```

```python
# What is the line of code doing? Input your own comment to explain.
# This command extracts the coefficients of the quadratic polynomial
inputs_quad
```

```python
# What is the line of code doing? Input your own comment to explain.
# output_quad_function is a polynomial function created using the coefficients from inputs_quad

output_quad_function = np.poly1d(inputs_quad)
```

```python
# What is the line of code doing? Input your own comment to explain.
# This line creates y-values for the quadratic model using the polynomial function
# output_quad_function is the polynomial object created earlier with np.poly1d
# Evaluating this function with our x values generates the predicted y values for our quadratic model
output_quad_exp = output_quad_function(x)
```

```python
# What is the line of code doing? Input your own comment to explain.

output_quad_exp
```

```python
# # What is the line of code doing? Input your own comment to explain.

plt.scatter(x,y, color = 'brown', label = 'data')
plt.plot(x,output_quad_exp, color = "pink", label = 'expected', linewidth = 3 )
plt.legend()
plt.show()
```

```python
# What if we try to set it to a function to the 6th power? Write your code here and comment it.

sixthpower = np.polyfit(x,y,6)
print(sixthpower)   
```

```python
# Name the polyfit output from the prior cell to 'poly_array'.

poly_array = np.poly1d(sixthpower)
```

```python
y_poly_exp = poly_array(x)
```

```python
# Does the curve fit the data or is it "over fit"?
# The curve fits the data very closely, but it may be overfitting
plt.scatter(x,y)
plt.plot(x,y_poly_exp)
plt.show()
```
